```python
def count_retry(log_filename):
    dat = open(log_filename).read().split('\n')
    result = 0
    for s in dat:
        if s.find('sent') >= 0 or s.find('resent') >= 0:
            result += 1
    return result

def count_throughput(log_filename):
    dat = open(log_filename).read().split('\n')
    result = 0
    for s in dat:
        if s.find('received data') >= 0:
            result += 1
    return result
```

```python
print count_retry('./log/try_250')
print count_throughput('./log/throughput_250')
```
